# Question 3: Plots and Discussion

```python
import numpy as np
import math
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

```python
# Actual data points
ns = list(range(1,16))
P_n = [.33, .5, .62, .7, .77, .81, .85, .88, .9, .92, .93, .95, .96, .96, .97]
```

## Weibull Model

```python
def weibull(a,c,n):
    return (1 - math.exp(-1*a*(n**c)))

a_hat = np.log(1/.67)
c_hat = np.log(1/a_hat)
weib_n = [weibull(a_hat,c_hat,n) for n in ns]

_ = plt.scatter(ns, P_n, color='red')
_ = plt.plot(ns, weib_n, color='blue')
_ = plt.legend(['Weibull predictions','Actual data'])
_ = plt.xticks(ns)
_ = plt.xlabel('Trial #n')
_ = plt.ylabel('$\hat{P}_n$')
plt.show()
```

## Replacement Model

```python
def replacement(theta, n):
    return (1 - (1 - theta)**n)

theta_hat = .33
repl_n = [replacement(theta_hat, n) for n in ns]

_ = plt.scatter(ns, P_n, color='red')
_ = plt.plot(ns, repl_n, color='blue')
_ = plt.legend(['Replacement model predictions','Actual data'])
_ = plt.xticks(ns)
_ = plt.xlabel('Trial #n')
_ = plt.ylabel('$\hat{P}_n$')
plt.show()
```

## Accumulation Model

```python
def accumulation(theta, n):
    return (n*theta)/(1 + n*theta)

theta_hat = .33/.67
accu_n = [accumulation(theta_hat,n) for n in ns]

_ = plt.scatter(ns, P_n, color='red')
_ = plt.plot(ns, accu_n, color='blue')
_ = plt.legend(['Accumulation model predictions','Actual data'])
_ = plt.xticks(ns)
_ = plt.xlabel('Trial #n')
_ = plt.ylabel('$\hat{P}_n$')
plt.show()
```

## Discussion of Results
These three plots show that they all do a somewhat good job in approximating the actual datapoints. If we especially look at the first few trials, all three are pretty accurate. Yet, as we examine the whole predicted distribution for three models, we see that:
<ul>
<li>Accumulation model, although performing well at first few iterations, fall short of displaying the actual variance of datapoints at further steps. This would create an issue in the long run, hence this is not the ideal model for the data we have. </li>
<li>Replacement model does a good job in approximating the data for first and last iterations among our 15 trials. Yet, as we look at the middle portion of the distribution, we see that the curve diverges from actual data points (especially in comparison with the Weibull distribution), implying that it does not perform well enough to be our model of choice.</li>
</ul>
Thus, we can conclude that Weibull distribution is the model that approximates our datapoints most accurately. At all 15 iterations, its divergence from the actual data is small enough to be considered reasonable. It makes sense that Weibull is our model of choice, since we know that it is a very flexible model. Tuning the c parameter may result in monotonically increasing, monotonically decreasing or peaked hazard, hence suitable to many situations.
