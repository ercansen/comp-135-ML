```python
import pandas as pd
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from skimage.feature import haar_like_feature_coord, haar_like_feature, draw_haar_like_feature
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

```python
X_train_df = pd.read_csv('data_digits_8_vs_9_noisy/x_train.csv')
y_train_df = pd.read_csv('data_digits_8_vs_9_noisy/y_train.csv')
X_test_df  = pd.read_csv('data_digits_8_vs_9_noisy/x_test.csv')
y_test_df  = pd.read_csv('data_digits_8_vs_9_noisy/y_test.csv')

X_train = X_train_df.values
y_train = y_train_df.values.ravel()
X_test  = X_test_df.values
y_test  = y_test_df.values.ravel()
```

```python
seed=12

scores = []
loglos = []
models = []

for i in range(1,41):
    logreg = linear_model.LogisticRegression(solver='liblinear', max_iter=i, random_state=12)
    logreg.fit(X_train, y_train)
    models.append(logreg)
    scores.append(logreg.score(X_test, y_test))
    loglos.append(metrics.log_loss(y_test, logreg.predict_proba(X_test)[:,1]))
    
    
```

```python
_ = plt.plot(range(1,41), scores, c='red')
_ = plt.xlabel('Number of iterations')
_ = plt.ylabel('Accuracy score')
_ = plt.title('Accuracy Score vs. Number of Iterations')
plt.savefig('score_vs_iter.png')
```

```python
_ = plt.plot(range(1,41), loglos, c='blue')
_ = plt.xlabel('Number of iterations')
_ = plt.ylabel('Log loss')
_ = plt.title('Log Loss vs. Number of Iterations')
plt.savefig('logloss_vs_iter.png')
```

```python
coefs = []

for i in range(40):
    coefs.append(models[i].coef_[:,0])
    
_ = plt.plot(range(1,41),coefs,color='purple')
_ = plt.xlabel('Number of iterations')
_ = plt.ylabel('Weight of pixel000')
_ = plt.title('Weight of Pixel000 vs. Number of Iterations')
plt.savefig('weight_pixel000_vs_iter.png')
```

```python
loglos_C = []
models_C = []

C_grid = np.logspace(-9,6,31)
for C in C_grid:
    logreg = linear_model.LogisticRegression(C=C, solver='liblinear')
    logreg.fit(X_train, y_train)
    models_C.append(logreg)
    loglos_C.append(metrics.log_loss(y_test,logreg.predict_proba(X_test)[:,1]))

_ = plt.plot(C_grid, loglos_C)
_ = plt.xscale('log')
_ = plt.xlabel('Values of parameter C (log scaled)')
_ = plt.ylabel('Log loss of model')
_ = plt.title('Log Loss vs. Parameter C')
plt.savefig('loglos_vs_C.png')
```

```python
print('Value of C that minimizes log loss: %.4f' % C_grid[np.argmin(loglos_C)])
print('Accuracy score of best model: %.4f' % models_C[np.argmin(loglos_C)].score(X_test,y_test))
```

```python
conf_m = pd.DataFrame(metrics.confusion_matrix(y_test,models_C[np.argmin(loglos_C)].predict(X_test)),columns=['Predicted 8','Predicted 9'], index=['Actual 8','Actual 9'])
conf_m.to_csv('confusion_matrix_1_3.csv')
```

```python
y_hat = models_C[np.argmin(loglos_C)].predict(X_test)
FN = (y_test > y_hat)
FP = (y_hat > y_test)
```

```python
fn_rows = X_test_df[FN].sample(n=9)
print(fn_rows.index)
fns = fn_rows.values

fp_rows = X_test_df[FP].sample(n=9)
print(fp_rows.index)
fps = fp_rows.values
```

```python
for i in range(9):
    _ = plt.imshow(fns[i].reshape(28,28), cmap='gray_r', vmin=0.0, vmax=1.0)
    plt.savefig('fn_'+str(i+1)+'.png')
```

```python
for i in range(9):
    _ = plt.imshow(fps[i].reshape(28,28), cmap='gray_r', vmin=0.0, vmax=1.0)
    plt.savefig('fp_'+str(i+1)+'.png')
```

```python
weights = models_C[np.argmin(loglos_C)].coef_.reshape(28,28)
_ = plt.imshow(weights, cmap='RdYlBu', vmin=-.5, vmax=.5)
plt.savefig('pixels.png')
```

```python
def efficient_rectangle_features(imm, px_wid):
    im = imm.copy()
    im = im.reshape(px_wid,px_wid)
    
    for i in range(px_wid):
        for j in range(1,px_wid):
            im[i][j] += im[i][j-1]
            
    for k in range(1,px_wid):
        im[k] += im[k-1]
        
    return im
```

```python
integral_images = np.empty((shoe_x_tr.shape[0], 28, 28))

for i in range(shoe_x_tr.shape[0]):
    integral_images[i] = efficient_rectangle_features(shoe_x_tr[i],28)
```

```python
feature_types = ['type-2-x', 'type-2-y']

feature_coord, feature_type = haar_like_feature_coord(width=integral_images.shape[2], height=integral_images.shape[1], 
                                                      feature_type=feature_types)
```
